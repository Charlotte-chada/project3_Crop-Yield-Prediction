```python
df3=pd.read_csv('data_3_crop.csv', na_values= ['n/a','\s+','N/A'], delimiter=',')
```

```python
#explore unique data in each column with unique method for df1
for col in df1:
    print(col, "::\n", df1[col].unique(), "\n\n")
```

```python
#explore unique data in each column with unique method for df2
for col in df2:
    print(col, "::\n", df2[col].unique(), "\n\n")
```

```python
#explore unique data in each column with unique method for df3
for col in df3:
    print(col, "::\n", df3[col].unique(), "\n\n")
```

```python
#convert data in df1 to 2 decimal

#convert Fahrenheit to Celsius
df1["Avg Min Temp *C"] = round((df1["Avg Min Temp *F"] - 32) * 5/9,2)

#convert inches to millimeters
df1["Yr Rain mm"] = round(df1["Yr Rain inches"] * 25.4,2)
```

```python
df1.head()
```

```python
# Drop unused columns
df1 = df1.drop(columns=["Avg Min Temp *F", "Yr Rain inches"])
```

```python
df1.head()
```

```python
#concat df1 and df2
df1_2 = pd.concat([df1,df2])
df1_2.shape
```

```python
df1_2.head()
```

```python
#merge data df1&df2 with df3
df=pd.merge(df1_2, df3, on=["Year", "Location"])
df.shape
```

```python
df.head()
```

```python
#explore unique data in each column in final df
#there are some issue : data entry issue, 
for col in df:
    print(col, "::\n", df[col].unique(), "\n\n")
```

```python
#fix data entry issue with replace method 
df["Heat Wave"]=df["Heat Wave"].replace('No','N')
```

```python
df["Dry Spell"]=df["Dry Spell"].replace({'n': 'N', 'no' : 'N', '0' : 'N'})
```

```python
df["Cold Wave"]=df["Cold Wave"].replace({'yes': 'Y', 'No' : 'N', 'n' : 'N','0' : 'N'})
```

```python
df["Wet Spell"]=df["Wet Spell"].replace({'No' : 'N', '1' : 'Y'})
```

```python
df["Irrigation"]=df["Irrigation"].replace({'yes' : 'Y', 'No' : 'N'})
```

```python
df["Crop Damage"]=df["Crop Damage"].replace({1. : 'Y', 0. : 'N'})
```

```python
for col in df:
    print(col, "::\n", df[col].unique(), "\n\n")
```

```python
#delete extra white space
df=df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
#detect null values
df.isnull().values.any()
df.isnull().sum()
```

```python
#check null columns and rows
print(df.isnull().any().sum()) #column
print(df.isnull().any(axis=1).sum()) #row
```
